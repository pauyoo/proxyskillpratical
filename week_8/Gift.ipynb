```python
def age(year):
    if year == 11:
        return "year 5"
    elif year == 12:
        return "year 7"
    elif year == 13:
        return "year 8"
    elif year ==  14:
        return "year 9"
    elif year == 15:
        return "year 10"
    else:
        return "year 10"

year = int(input("What is your age: "))
print(age(year))
```
