```python
def check_number(value):
    if value > 0:
        return "Positive"
    elif value < 0:
        return "Negative"
    else:
        return "Zero"
print  (check_number(5))
user_input = int(input("Enter a number: "))
result = check_number(user_input)
print(f"The number is {result}")
```
