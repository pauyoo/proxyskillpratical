```python
## 4. Calculator Program

operator = input("Choose an operator(+ - * /)")
num1 = float(input('Enter your first number'))
num2 = float(input('Enter your first number'))

if operator == '+':
    result = num1 + num2
    print(f"This is the result :{result} ")
elif operator == '-':
    result = num1 - num2
    print(f"This is the result :{result} ")
elif operator == '*':
    result = num1 * num2
    print(f"This is the result :{result} ")
elif operator == '/':
    result = num1/ num2
    print(f"This is the result :{result} ")
```

```python
## 5. Weight Conversion
weight = float(input("Enter your weight:"))
unit = input('Kilograms or Pounds? (K or L):')

if unit == 'k':
    weight= weight * 2.205
    unit = "Lbs."
    print(f"This is your weight: {round(weight, 1)} {unit}")
elif unit == 'L':
    weight = weight / 2.205
    unit = "Lbs."
    print(f"This is your weight: {round(weight, 1)} {unit}")
else:
    print(f" {unit} was not valid")
```
