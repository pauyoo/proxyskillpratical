```python
## CGPA calculator

score = float(input("what is your score"))

if score >= 4.49:
    grade = "First Class"
elif score >= 3.49:
    grade = "2nd Class Upper"
elif score >= 2.49:
    grade = "2nd Class Lower"
else:
    grade = "3rd class"
print("Grade:", grade)
```
