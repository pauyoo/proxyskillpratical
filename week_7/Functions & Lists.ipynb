In Python, a function is a block of reusable block of code that performs a specific task.
You define a function using the def keyword, followed by the function name, parameters (optional), and a colon. The function body is indented.

invoke funct: Calling out the function 



```python
def happy_birthday():
    print("Happy birthday")
    print("How old are you")
    print("Happy birthday to you")

happy_birthday()
```

```python
## Passing arguement to the function using f string

def happy_birthday(my_name):
    print(f"Happy birthday to {my_name}")
    print("How old are you")
    print("Happy birthday to you")
    print()

happy_birthday("Gideon")
happy_birthday("Noah")
happy_birthday("Susan")
```

```python
## Passing arguement and mutiple parameters to the function

def happy_birthday(my_name, age):
    print(f"Happy birthday to {my_name}")
    print(f"You are {age} years old")
    print("Happy birthday to you")
    print()

happy_birthday("Gideon", 14)
happy_birthday("Noah", 14)
happy_birthday("Susan", 15)



```

Difference btw LIST, SET, & TUPLE
list: [] ordered and changeable. Duplicates OK 
Set: {} Unordered and immutable but add/remove OK. No duplicates
Tuple: () Ordered and changeable. Duplicates ok and faster.
