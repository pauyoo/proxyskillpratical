```python
def grade(score):
    if 90 <= score <= 100:
        return "Grade A+"
    elif 80 <= score < 90:
        return "Grade A"
    elif 75 <= score < 80:
        return "Grade B+"
    elif 60 <= score < 75:
        return "Grade B"
    elif 55 < score < 60:
        return "Grade C"
    else:
        return "Grade F"

score = int(input("What is your Score: "))
print(grade(score))
```
