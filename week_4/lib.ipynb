```python
## Goals scored by different clubs 
data = {
    "Clubs": ["Man United", "Chelsea", "Arsenal", "Liverpool"],
    "Goals Scored": [65, 90, 72, 80]
}

df = pd.DataFrame(data)
print(df)
```

```python
average_goals = df["Goals Scored"].mean()
print("Average Goals:", average_goals)
```

```python
top_club = df.loc[df["Goals Scored"].idxmax()]
print("Top Club:", top_club["Clubs"])
```

```python
plt.bar(df["Clubs"], df["Goals Scored"])
plt.title("Football Club Goals")
plt.xlabel("Clubs")
plt.ylabel("Goals Scored")

plt.savefig("football_goals.png", dpi=300)
plt.show()
```

```python
import pandas as pd

data = {
    "Clubs": ["Man United", "Chelsea", "Arsenal", "Liverpool"],
    "Goals Scored": [65, 90, 72, 80],
    "Goals Against": [40, 45, 38, 30]
}

df = pd.DataFrame(data)
print(df)
```

```python
## To calculate the goal difference
df["Goal Difference"] = df["Goals Scored"] - df["Goals Against"]

print(df)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Clubs"], df["Goal Difference"])
plt.title("Goal Difference by Club")
plt.xlabel("Clubs")
plt.ylabel("Goal Difference")

plt.savefig("goal_difference.png", dpi=300)
plt.show()
```

```python
## To sort teams by goal difference 
df_sorted = df.sort_values(by="Goal Difference", ascending=False)
print(df_sorted)
```
